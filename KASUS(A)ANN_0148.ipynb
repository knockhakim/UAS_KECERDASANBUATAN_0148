```python
!pip install keras scikit-learn
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.metrics import classification_report
```

```python

# Membaca dataset
df = pd.read_csv('iris.csv')
```

```python
# Menampilkan informasi dataset
print("Informasi Dataset Iris:")
print("======================")
print("Jumlah Sampel:", len(df))
print("Jumlah Fitur:", len(df.columns) - 1)  # Mengurangi 1 untuk kolom kelas/target
print("Jumlah Kelas:", len(df['species'].unique()))
print("======================")
print(df.info())
```

```python


# Mengambil fitur dan label
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values


```

```python

# Melakukan encoding pada label
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
# Membagi dataset menjadi data training dan data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
# Menginisialisasi model ANN
model = Sequential()
```

```python
# Menambahkan input layer dan hidden layer_1
model.add(Dense(16, input_shape=(4,), activation='relu'))
```

```python
# Menambahkan hidden layer_2
model.add(Dense(16, activation='relu'))
```

```python

# Menambahkan hidden layer_3
model.add(Dense(16, activation='relu'))
```

```python
# Menambahkan hidden layer_4
model.add(Dense(16, activation='relu'))
```

```python

# Menambahkan output layer
model.add(Dense(3, activation='softmax'))
```

```python
# Mengompilasi model dengan optimizer Adam
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Melatih model menggunakan data training
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
```

```python
# Memprediksi kelas untuk data testing
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
```

```python
# Membuat classification report
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
```

```python
# Menampilkan classification report
print("Classification Report:")
print(report)
```

```python
# Evaluasi model menggunakan data testing
loss, accuracy = model.evaluate(X_test, y_test)
print("Loss:", loss)
print("Accuracy:", accuracy)
```
